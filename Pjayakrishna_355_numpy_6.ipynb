### `Where()`
- np.where() is a function in NumPy that returns the index of elements in an array that satisfy a given condition.
- `np.where(condition, [x, y])`



```python
import numpy as np,pandas as pd
```

```python
arr = np.array([1, 7, 3, 9, 2, 6])
indices = np.where(arr > 5)
print("Original array:", arr)
print("Indices where elements are greater than 5:", indices)
print("Elements greater than 5:", arr[indices])
```

```python
arr = np.random.randint(-5,10,(3,5))# creating an array with elements ranging from -5 to 9
arr
```

```python
 # here the condition is given which returns indices of elements that are less than 0, this indices are given to the array and value is set to 0
arr[np.where(arr<0)] = 0
print(f'Array after all the negative values are replaced by 0 \n{arr}')
```
